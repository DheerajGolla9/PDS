```python
import pandas as pd
import numpy as np
```

# Step 1: Load the dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
```

# Displaying the first few rows of the DataFrame

```python
print("Original DataFrame (first 10 rows):")
df.head(10)
```

# Printing the data types to verify the column types


```python
print("Missing values before handling:")
print(df.isnull().sum())
```

```python
print("Data types of each column:")
df.dtypes
```

# Step 2: Select numeric columns


```python
numeric_df = df.select_dtypes(include=np.number).copy()
```

# Verifying the data types of numeric columns


```python
print("Data types of numeric columns:")
numeric_df.dtypes
```

# Printing the numeric columns

```python
print(numeric_df.head(10))
```

# Step 3: Calculating the mean of each numeric column


```python
mean_values = numeric_df.mean()
```

# Printing the calculated mean values for each numeric column


```python
print("Calculated mean values for each numeric column:")
print(mean_values)
```

# Step 4: Filling missing values with the mean


```python
numeric_df.fillna(mean_values, inplace=True)
```

# Printing the DataFrame after replacing missing values with the mean


```python
print("Numeric DataFrame after filling missing values with mean:")
numeric_df.head(10)
```

```python
print(numeric_df.dtypes)
```

# Step 5: Updating the original DataFrame with the modified numeric columns


```python
df[numeric_df.columns] = numeric_df
df.head(10)
```

# Step 6: Selecting and updating non-numeric columns to Original DataFrame


```python
boolean_cols = df.select_dtypes(include=['object', 'bool']).columns
df[boolean_cols] = df[boolean_cols].apply(lambda col: col.fillna(col.mode()[0]).astype(str))
```

#Verifying the Missing values after handling

```python
print("Missing values after handling:")
print(df.isnull().sum())
```
