```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("/content/train.csv")
df.head()
```

```python
print("Initial dataset shape:", df.shape)
print("Missing values:\n", df.isnull().sum())
```

# Drop 'New_Price' due to high number of missing values


```python
df.drop(columns=['New_Price'], inplace=True)
```

# Fill missing values


```python
df['Mileage'].fillna(df['Mileage'].mode()[0], inplace=True)
df['Engine'].fillna(df['Engine'].mode()[0], inplace=True)
df['Power'].fillna(df['Power'].mode()[0], inplace=True)
df['Seats'].fillna(df['Seats'].median(), inplace=True)
```

```python
print("Missing values:\n", df.isnull().sum())
```

# Remove units (e.g., "kmpl", "CC", "bhp") and convert to float


```python
df['Mileage'] = df['Mileage'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
df['Engine'] = df['Engine'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
df['Power'] = df['Power'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
```

# One-hot encode categorical variables


```python
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'], drop_first=True)
df
```

# Create a new derived feature: Car Age


```python
current_year = datetime.now().year
df['Car_Age'] = current_year - df['Year']
df.head()
```

# Select specific columns


```python
selected_df = df[['Name', 'Location', 'Year', 'Car_Age', 'Mileage', 'Engine', 'Power', 'Price']]
```

# Filter cars with mileage greater than 20 kmpl


```python
filtered_df = selected_df[selected_df['Mileage'] > 20]
```

# Rename column: Power â†’ Horsepower


```python
renamed_df = filtered_df.rename(columns={'Power': 'Horsepower'})
```

# Mutate: Create Mileage per Year (avoid division by zero with +1)


```python
renamed_df['Mileage_per_Year'] = renamed_df['Mileage'] / (renamed_df['Car_Age'] + 1)
```

# Arrange: Sort by Price descending


```python
arranged_df = renamed_df.sort_values(by='Price', ascending=False)
```

```python
print("\nTop expensive cars (filtered):")
print(arranged_df.head())
```

# Summarize: Average price and total cars per location


```python
grouped_summary = df.groupby('Location')['Price'].agg(['mean', 'count']).reset_index()
grouped_summary.rename(columns={'mean': 'Avg_Price', 'count': 'Total_Cars'}, inplace=True)
```


# Displaying final results


```python
print("\nAverage price and count per location:")
print(grouped_summary.head())
```
